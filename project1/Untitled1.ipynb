```python
import numpy as np
import pandas as pd
import math
import scipy
import sklearn
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn import impute
from sklearn import model_selection
from sklearn.svm import SVR
from sklearn.preprocessing import QuantileTransformer
from sklearn.neighbors import LocalOutlierFactor
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
# Read file
train_x = pd.read_csv(
    "X_train.csv",
    sep = ",",
    index_col = "id"
)

train_y = pd.read_csv(
    "y_train.csv",
    sep = ",",
    index_col = "id"
)

test_x = pd.read_csv(
    "X_test.csv",
    sep = ",",
    index_col = "id"
)
```

```python
train_x
```

```python
def mp(x):
  if x == 1:
    return True
  else:
    return False
```

```python
qt = QuantileTransformer(output_distribution="normal", random_state=np.random.RandomState(304))

impute_data = train_x.fillna(train_x.median())
clf = LocalOutlierFactor()
outliers = clf.fit_predict(impute_data)
liers = np.array(list(map(mp, outliers)))
impute_data = impute_data[liers]
labels = train_y[liers]
selecter = SelectKBest(f_regression, k=20)
impute_data = selecter.fit_transform(impute_data, labels.values.flatten())
final_x = qt.fit_transform(impute_data)



```

```python
impute_test = test_x.fillna(train_x.median())
impute_test = selecter.transform(impute_test)
final_test = qt.transform(impute_test)
```
